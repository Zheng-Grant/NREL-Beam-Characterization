# For visualizing the difference between processing methods

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
data_folder = r"C:\Users\qzheng\Documents\bcs_results"
energy_90 = os.path.join(data_folder, "result_centered_90_energy_original", "NREL_aggregated_results_90_energy.csv")
energy_60 = os.path.join(data_folder, "result_centered_60_energy_original", "NREL_aggregated_results_60_energy.csv")

# First column is file name, second column is X position, third column is Y position. First row is header.
energy_90_df = pd.read_csv(energy_90, header=0, names=["file_name", "x_position", "y_position"])
energy_60_df = pd.read_csv(energy_60, header=0, names=["file_name", "x_position", "y_position"])

energy_90_x = energy_90_df["x_position"].values
energy_90_y = energy_90_df["y_position"].values
energy_60_x = energy_60_df["x_position"].values
energy_60_y = energy_60_df["y_position"].values

# crate 1x2 subplots of histograms of energy_90_x and energy_60_x, and energy_90_y and energy_60_y
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
bin_counts = 50
axs[0].hist(energy_90_x, bins=bin_counts, alpha=0.5, label="Energy 90")
axs[0].hist(energy_60_x, bins=bin_counts, alpha=0.5, label="Energy 60")
axs[0].set_title("X Position Distribution")
axs[0].set_xlabel("X Position")
axs[0].set_ylabel("Frequency")

axs[1].hist(energy_90_y, bins=bin_counts, alpha=0.5, label="Energy 90")
axs[1].hist(energy_60_y, bins=bin_counts, alpha=0.5, label="Energy 60")
axs[1].set_title("Y Position Distribution")
axs[1].set_xlabel("Y Position")
axs[1].set_ylabel("Frequency")

axs[0].legend()
axs[1].legend()
plt.tight_layout()
```
