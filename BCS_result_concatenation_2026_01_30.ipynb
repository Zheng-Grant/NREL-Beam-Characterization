```python
import numpy as np
import os
from glob import glob
import pandas as pd

result_base_folder = r"C:\Users\qzheng\Documents\bcs_results\result_centered"

buffer_dataframe = pd.DataFrame(columns=["relative_path", "Centroid_X", "Centroid_Y"])

level_1_folders = os.listdir(result_base_folder)
for level_1_folder in level_1_folders:
    level_1_folder_path = os.path.join(result_base_folder, level_1_folder)
    level_2_folders = os.listdir(level_1_folder_path)
    for level_2_folder in level_2_folders:
        level_2_folder_path = os.path.join(level_1_folder_path, level_2_folder)
        level_3_folders = os.listdir(level_2_folder_path)
        for level_3_folder in level_3_folders:
            level_3_folder_path = os.path.join(level_2_folder_path, level_3_folder)
            level_4_folders = os.listdir(level_3_folder_path)
            for level_4_folder in level_4_folders:
                level_4_folder_path = os.path.join(level_3_folder_path, level_4_folder)
                result_csv_file = glob(os.path.join(level_4_folder_path, "*.csv"))
                assert len(result_csv_file) == 1, f"Expected exactly one CSV file in {level_4_folder_path}, but found {len(result_csv_file)}"
                result_csv_file_path = result_csv_file[0]
                # First row is header
                data = pd.read_csv(result_csv_file_path, header=0)
                # concatenate to the buffer dataframe
                buffer_dataframe = pd.concat([buffer_dataframe, data], ignore_index=True)


# Save the buffer dataframe to a new CSV file
output_csv_file_path = os.path.join(result_base_folder, "aggregated_results.csv")
buffer_dataframe.to_csv(output_csv_file_path, index=False)
print(f"Aggregated results saved to {output_csv_file_path}")




```
